## Replace Pronouns with Coreference Model



```python

# Flatten document into lists of token
def flatten(line):
    out = []
    for item in line:
        if isinstance(item, list):
            out.extend(flatten(item))
        else:
            out.append(item)
    return out

# Stringitize list of tokens back into document
def stringitize(txtlst):
    string = flatten(txtlst)
    string = ' '.join(str(e.strip()) for e in string)
    return string

def repl_pron_prop(text):
  # pre-step: tokenize and cluster with AllenNLP
  # code from:  https://code.likeagirl.io/obtaining-allennlp-coreference-resolution-readable-clusters-in-python-37cd964d6ca0
  pred = predictor.predict(document = text)
  clusters = pred['clusters']
  tokenlst = pred['document'] 

# STEP 1: Align POS with all words in the text.  
# Use token list defined by AllenNLP to align, but these must be equal.
  """Output is a list of tuples: (POS, token)"""
  doc = nlp(text) # Assigns text to spaCy
  poss = [tok.tag_ for tok in doc]
  if len(tokenlst) == len(poss):
      posntuple = list(zip(poss, tokenlst))
  else: 
    print("document and POS tokens are different lengths")
    posntuple = None

  # STEP 2: Align POS with Words by cluster
  # flatten the clusters to show the 
  flclust = []
  for i in range(len(clusters)):
    flatter = list(sorted(set(flatten(clusters[i])))) # ensure the list is unique
    flclust.append(flatter)
  # assign words and POS tags to clusters
  wordclust = []
  for cl in flclust:
    wrdclust = []
    for c in cl:
      wrdclust.append(posntuple[c])
    wordclust.append(wrdclust)

  # STEP 3: Find the right proper noun to replace pronouns with
  # if the cluster contains NNP or NNPS, or DT immediately prior to NN or NNS, or just NN or NNS:
  # assign the Proper Noun in this order.
  propnoun_perclust = []
  for wc in wordclust:
    dt = None
    proper = None
    for w in wc:
      if proper != None: continue
      elif (w[0] == 'NNP' or w[0] == 'NNPS'): proper = w[1]
    for w in wc:
      if proper != None: continue
      elif w[0] == 'DT': dt = w[1]
    for w in wc:
      if proper != None: continue
      elif (w[0] == 'NN' or w[0] == 'NNS') and dt != None: proper = w[1]
    for w in wc:
      if proper != None: continue
      elif w[0] == 'NN' or w[0] == 'NNS': proper = w[1]
    propnoun_perclust.append(proper) #add the cluster's proper noun to the list; may be None.
  if len(propnoun_perclust) != len(clusters): 
    print("ERROR: not enough proper nouns for the clusters.")
    new_document = None
    return new_document # Returns blank document if error

  # STEP 4: map pronouns in the tokenized document to the proper noun identified
  # print statements have been commented out, but can be uncommented for inspection
  propnoun_mapper = {}
  for i in range(len(propnoun_perclust)):
    if propnoun_perclust[i] == None: continue 
    count_pronoun = 0
    for c in flclust[i]: # for each token in the cluster  
      if posntuple[c][0][:3] == "PRP":  # only if the token is a pronoun
      # Removing the i, me, us and you, because the context of this is speech discourse, where the direct communicants are often unacknolwedged
        if (posntuple[c][1].lower() != "i") and (posntuple[c][1].lower() != "me") and (posntuple[c][1].lower() != "us") and (posntuple[c][1].lower() != "you"):
        #### Indented ####
          #print(f'\nAssign Proper Noun:\n{propnoun_perclust[i]} \nTo the pronoun(s):')
          count_pronoun += 1
          #print(posntuple[c][1])
          propnoun_mapper[c] = propnoun_perclust[i] # map the identified proper noun to the index
  # STEP 5: apply the map to the tokenized document
  new_document = []
  count_mapper = 0
  for i in range(len(tokenlst)):   
    if i in propnoun_mapper: 
      new_document.append(propnoun_mapper[i])
      count_mapper +=1
    else: new_document.append(tokenlst[i])
  ratio = count_mapper/len(tokenlst)
  print(f'\nFINAL count of replaced words: ', count_mapper)
  print(f'TOTAL size of tokenized text: ', len(tokenlst))
  print(f'RATIO: ', ratio)
  new_document = stringitize(new_document)

  return(new_document, count_mapper, len(tokenlst), ratio)
```

### Replace pronouns across entire corpus

```python
i = 0
coref = []
replaced = []
tokens = []
ratio = []

# iterate through dataframe, printing out each index to monitor progress
for index, value in df['speech_final'].items():
    print(f"~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n{i} of 87, Index: {index}")
    a, b, c, d = repl_pron_prop(value)
    coref.append(a)
    replaced.append(b)
    tokens.append(c)
    ratio.append(d)
    i+=1

df['Final_CoRef']  = coref
df['Count_Replaced'] = replaced
df['Count_Tokens'] = tokens
df['Ratio_Replaced'] = ratio
```
